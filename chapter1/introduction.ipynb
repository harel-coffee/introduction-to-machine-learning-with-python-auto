```python
import seaborn as sns
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
iris_dataset
```

```python
print("Keys of iris dataset: %s" % iris_dataset.keys())
print("Features of iris_dataset: %s" % iris_dataset.feature_names)
print('Data shape: %s' % (iris_dataset.data.shape,))
print('Targets: %s' % (iris_dataset.target_names,))
```

```python
data = iris_dataset.data
target = iris_dataset.target
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(data, target, random_state=0)
print("Train shape: %s" % (X_train.shape,))
print("Test shape: %s" % (X_test.shape,))
```

```python
df=DataFrame(data, columns=iris_dataset.feature_names)
df['target']=target
sns.pairplot(df, hue='target')
sns.pairplot?
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, Y_train)
```

```python
Y_pred=knn.predict(X_test)
Y_pred
```

```python
print("Test set score: %s" % (Y_pred==Y_test).mean())
print("Test set score by knn: %s" % knn.score(X_test, Y_test))
```
